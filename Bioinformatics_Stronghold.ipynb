#### Q2 Transcribing DNA to RNA

Given a DNA string t corresponding to a coding strand, its transcribed RNA string u is formed by replacing all occurrences of 'T' in t with 'U' in u.

**Given:** A DNA string t having length at most 1000 nt.

**Return:** The transcribed RNA string of t.

```python
def Transc(DNA):
    RNA = ''
    DNA = DNA.upper()
    for i in DNA:
        if i == 'T':
            RNA += 'U'
        else:
            RNA += i
    return RNA

#Test
Transc('GATGGAACTTGACTACGTAAATT')
Transc('AGGCGATTCTTAGTTCTACCCGGCCGAATTTCGTCTACGCCTCAATCGCTGGGTAAGATCGGCAGAGTGTGACCACGGTCTGGCCCCCTTAGCGTATCGGGATTTATGAGCCCATAGCCCTCATACGAGGCGCCGGTATCTCCAGACAAGCTGTGGAGCCACCAGTCAGTAATGAACGTTCGCAGGAAATCAGTCCAGTAAACTGGTTGACGAGCTTAAGCGCTCGGGGGCAAAATCGAGTACGCGAATAGAATCGACACCTGATTGTAGCCTCTCCTACGGGCCGACGATAAAGTTATTTCTAGCTCTTTTGGGTGGATAAAAGCCGTACCGCCAATTAGGTTAGCCCCAGAGCGAGCTACTGAGCCTCAGGCGCACACCCTTCGGTTTTAGAGTCACGTTGAAGCAATCGCATCATGCATTACATCATTAATGAGGGAATACATGGCTGATTTGTTATGCGACCGAACCTCCGCGGTGTCGCACCGCTTAGGTGAGAACTCCACTAGGCCCTCGATGGGAGATAAATAGATAAGAAGTCGGTCAAAGCATTGTTTCCCTGGCTAAGTAGCAGGGCCTGTTTCGACAACGGATAACCTACTCCATGATAGCAATAATGCCGAAATGCGACAAATCTTCCTTGTCACGCTCGACGCGTGCGGTGGCGGGACTCGCGACTGATAACTTTAAGTTCGGTACCTCTAGGGTTCCCCTAAGACCGCCGACAATAATGAAGGAGGGTGTTTCCTATCCTGTAGACTGTGCCGGTGGTAGCACGGCATTTCCCCAGCATTCAGCCAATGCTTTTTACTTCCTACAATAATTGTCCTCCTTGTGCGACCAATAAGGAGTTATGCTGCTCAAAGACTACGACACCCGCCGGACGGGGTTCCTCATAATAATCGTAGCGACGAAATCTATCTACCAGCCATTTTCCTCTCGATCCAGCGGGCGAACCGCAGGAGGATAGCAAACTGCCGACCTGTTCTCATGTT')
```

#### Q3 Reverse Complement

**Given:** A DNA string s of length at most 1000 bp.

**Return:** The reverse complement sc of s.


```python
def RevComp(DNA):
    DNA = DNA[::-1]
    DNA_rc = ''
    for i in DNA:
        if i == "A":
            DNA_rc += 'T'
        elif i == "T":
            DNA_rc += 'A'
        elif i == "G":
            DNA_rc += 'C'
        elif i == "C":
            DNA_rc += 'G'
    return DNA_rc

#Test

RevComp('AAAACCCGGT')
RevComp('GGAATGCATCTCTATACAGTATCTCTCCGGACCTCAGCCGATTAGTGTAACAAAATCATAAGACGCCCAGGAATTTTTCTGTAAAAGTGTAAGTTCCTATGGCGCTTTAAGGCGATATACCTCGAATTGTTGCATGCTAAACACGAGGCTTGGTTAATGATCCATTAGTGATGAAAGTTGGCGACTCAAATTCTAGTGTTAACAACACTCCAGAGAGTATCCGCTTTTAGCCTTATCATAGATCTACGCTATTTATTCCCACTAAATCATCGTGACAACGCATGCTCTAGCCCGTATATGTGCAGGTTCACAAGTCGTGGACGAGAACGCGTGAAAAATGTCCTTCTGTCACCCAAATGGCCTAAGGTTAACGGAGTCACTTTGAAGCTAATCTTTGCTCTCGCCATCTATCAGTGGGAACCGGTAAAATAACGAGCAAACAGGGCCCTTATGCCGGATTTGTCGGCCAATGGGCACAGTGCACCACATACCATTGACTATCAAATCAGGACACTATGCTCAACCTCATAGACTCGCGCAGAATTATCCCGCCAGACGTCATACTTCTCGCCATAAAGTAGGGTTGAAGGCGAGCCGCCGGGCGTAAACAAGTCGGCGTAATCAAACCCAGTATGCCCATACGCCCCAGTGCCGAGTCTCGGGAGACCATACTAATTCCTTGCTGTTTCGGAAAGTCTCTTGAGTCGAAGATACCAGTTAAGTGTAGGCACACTCAGCTCACGACACCCTCCATTCTGCCCTGTGGCCATGGAGGTTGCGCTTGTTACTGTTCTGAAGCTCAAACAATCCTTTCCAGTCCGGGTAATGTTCCAGGGACCTTCCCATGGCAGCCCATTTGCGTGACGTTCGTTTCCGGG')
```

#### Q4 Rabbits and Reccurences 

**Given:** Positive integers n≤40 and k≤5.

**Return:** The total number of rabbit pairs that will be present after n months, if we begin with 1 pair and in each generation, every pair of reproduction-age rabbits produces a litter of k rabbit pairs (instead of only 1 pair).

```python
def WascallyWabbits(n,k):
    '''
    n is number of months
    k is number of offsprings after 1 month 
    '''
    
    if n == 1:
        return 1
    elif n ==2:
        return k
    
    oneGen = WascallyWabbits(n-1,k)
    twoGen = WascallyWabbits(n-2,k)
    
    if n <=4:
        return (oneGen + twoGen)
    
    return (oneGen + twoGen * k)

WascallyWabbits(28,2)
```

#### Q6 Counting Point mutations

**Return:** The Hamming distance dH(s,t).

Given two strings s and t of equal length, the Hamming distance between s and t, denoted dH(s,t), is the number of corresponding symbols that differ in s and t.

```python
def HammingD(A,B):
    HamDis = 0
    for i in range(len(A)):
        if A[i] != B[i]:
            HamDis += 1
    return HamDis

#Test
HammingD('GAGCCTACTAACGGGAT','CATCGTAATGACGGCCT')
HammingD('GCTCCACCAGGGGTATAATACTGAGCTGTCCGACCCTAGAAACGGATCTTGGCGGTACTTATATGTTAGTTTAGAGATTACAAAAGGATTCCCCTGGGTCCAATTTTCGCACACGCCTATAGTCCTATATATCCCGGAAGTTCTTCTCAATCGTGGGGGACGGGAGTGATTCGGGTCGGAGCATATTACGTTGGTCATAACTGATTGAAAGGTCAACAGTGGCATCCGAGCGGCCCCCCTAGTTACGTTGCTCTCGGCACCCCACTCGCCGAGCCCCTACCAAGATGAAGTGCGGGTACTAAAGTGGCTCAACTCATGGGTGTCGCGCTCAAGCTCGAGGCCCGCATTAACCCACTGTTAGGCTGCCGTTCGGTCGACTTCGAAGGAATAGGTGCATCCCGGGTATTTCTGAACGGTTGGAGCACTTCGGCTTAATGTCAAGCCGCGTTCAGTAGTTTTCGGGCGCTGCCCCAATCAGGCGTCACGATGTCGGTTGAGGTTAACTATAGTTTGCGGGTAAATCCTCTCAGAAAACTATTCTACCTCGCTCAACCTAGGGAATACCCCCCGACTGAGGTTCACAGTTCATCGGCACTAAGTCCCAGATAGGGGGAGGGTATCAAATTGAACCTTGTCTTTCACCGTTCACGGAAGCACACTATCGGTATTCATTGCCTGACTAGAGCTAGCGTGCAGACGCGAACCGGTATCATCGCAGAAGGTCGCACCTCTGCTGTCACGTACTTTCGAACTAGAGACGCATGTACGTTTATCGTCTTCCGCCTAAGCCCGGCAAACCTCTATTCCGTGTGGGTATCTCCTATGCTCTCGAAGGGAAGGTGAGTCCGAGCCCTTCCTGTATAGAACACCTCCCTTAATTGCCCGAGCTCTCTCTTGTCGAACCATTTAACTTCCCGTGGCTGTGCTCAGGCGCGTCAGGGCGTGACTAC','GACCGTTCAGCTGCGCAGTAGACAGGTCCTCCACCCTGGAATCACATCTGTACGATATGTACATGTTAGTAACGTTAAGACAATACGCATCGTTTGGGTAGGATAATCGTTCGAGTCGAAGGGTTGATATAACCCATACGGGCACCTCAACAGTTGAGCATAGGAATGTGACCCCTGATCTCATACTACGGTAGGCAAACCAGATCGTTAGGGCGGGTTTAGGGTCCCCCACCCCATACTGGGTACTTAGGTTTTGACAAACGTATTGATGAGGCCGTGACAAGATCACGTGATGGTACGTAAGACAGACCTCGATTGCGCATGGTGGTCAATCCAAACACCTTGTGTGAAACACTTGTCCGCAGTCATCATATTGTCATTGAAGAAGGTGTTGCACCGCGGGTAAAGCTTTCGGGTTGGCCCAATCCATGAGTATAGCCAGCAAGATCCAGTATTGTGAGAGCACTGACCCCTACGGACGATAAGAGCTGGCATTAGGTTGACCACTTTTTGGAGCTATATCATCGCGGTATGTCTTCATGTATTGCAGCACCTTCATCATAGCCCCCAACGAATAACTATGGATCATCGAATCTGTGCCACAGAATATGTCATTCCGTTTAATCGCCCCTGTTCATCCACTGGCTAATGAAAAATACGTAAGTCGATCTTCGCCATATTTCTAGTCCAAGCGAATGGCGAGACGTGACGAACGGAGAAGGGGGCAACACTGAATAGACCTACCTTCGAGCAAAATCCGGCTGACCGATCGTAGGCGTCAAACACAGCCGGTCCCGCATCCGTCGGGGCAGTCTCGCGCCGAAGAACAGCATTCGTAGGGGAGCCGGATCCGGTACTGATTATGCCACCTGCTTTTAACGTCCCACCTATCTTCTACCGCCGCCATCTGGTGCTGGTGGTTGATCGCTGTTTGGTAAGGGATGCGACCC')
```

#### Q7 Mendel's First law (Probability)

***Given:*** Three positive integers k, m, and n, representing a population containing k+m+n organisms: k individuals are homozygous dominant for a factor, m are heterozygous, and n are homozygous recessive.

***Return:*** The probability that two randomly selected mating organisms will produce an individual possessing a dominant allele (and thus displaying the dominant phenotype). Assume that any two organisms can mate.

```python
from scipy.special import comb
# scipy.special.comb this if for combination, similarly there is a perm for permutation
def DomAllele(k,m,n):
    '''
    k = AA 
    m = Aa
    n = aa
    totalpop is the total number of animals in ecosystem
    totalComb is the total combinations possible outcomes of all animals mating 
    DomComb is the total combinations with Dominant phenotype i.e. AA or Aa
    probab is the probabilty of having a Dominant off spring is 2 random animals mated
    
    Reasoning for DomCom formula is that it is sum of all these outcomes
    comb(k,2) : all combination of AA will be dominant
    k*m and k*n : any mating of Aa and aa with AA will be dominant
    0.5*m*n : as 50% of offsprings of mating between Aa and aa are dominant
    0.75*comb(m,2) : as 0.75 of the off springs of Aa mating with Aa will be dominant
    refer to https://stackoverflow.com/questions/25119106/rosalind-mendels-first-law-iprb
    '''
    totalPop = k + m + n
    totalComb = comb(totalPop,2)
    DomComb = comb(k,2) + k*m + k*n + 0.5*m*n + 0.75*comb(m,2)
    probab = DomComb/totalComb
    return probab

DomAllele(2,2,2)
DomAllele(30,28,29)
```
